```python
#
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

#
img = cv2.imread('imori.jpg')
img_noise = cv2.imread('imori_noise.jpg')
#
img5 = img.copy()
#RGB->HSV
hsv = cv2.cvtColor(img5,cv2.COLOR_RGB2HSV)
#HSV
hsv[...,0] = (hsv[...,0]+180)%360
#HSV->RGB
img5 = cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)
#
cv2_imshow(img5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
