```python
#ライブラリをインポート
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

#画像の読み込み
img = cv2.imread('imori.jpg')
img_noise = cv2.imread('imori_noise.jpg')
#処理
img5 = img.copy()
#RGB->HSV変換
hsv = cv2.cvtColor(img5,cv2.COLOR_RGB2HSV)
#HSV反転
hsv[...,0] = (hsv[...,0]+180)%360
#HSV->RGB変換
img5 = cv2.cvtColor(hsv,cv2.COLOR_HSV2RGB)
#画像を表示する
cv2_imshow(img5)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
