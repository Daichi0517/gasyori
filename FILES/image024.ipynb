```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
#画像の読み取り
img = cv2.imread('imori.jpg')
#画像のサイズを変更
out = cv2.resize(
    img, (int(img.shape[1]*1.5), int(img.shape[0]*1.5)), interpolation=cv2.INTER_NEAREST)
#画像の表示
cv2_imshow(out)
```
