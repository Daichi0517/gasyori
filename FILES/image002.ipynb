```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import requests
from io import BytesIO


# ここで処理
def processing(img):
    b = img[:, :, 0].copy()
    g = img[:, :, 1].copy()
    r = img[:, :, 2].copy()

    out = 0.2126*r + 0.7152*g + 0.0722*b
    out = out.astype(np.uint8)
    return out

# 画像の読み込み
imori_img = cv2.imread('imori.jpg')
answer_img = imori_img.copy()
answer_img = processing(answer_img)

# サブプロットを作成して、2つの画像を並べて表示
plt.figure(figsize=(10, 5))  # サイズを指定

# オリジナルの画像を表示
plt.subplot(1, 2, 1)  # 1行2列の1番目のプロットに表示
plt.imshow(cv2.cvtColor(imori_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original Image')

# 答えの画像を表示
plt.subplot(1, 2, 2)  # 1行2列の2番目のプロットに表示
plt.imshow(cv2.cvtColor(answer_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Answer Image')

plt.show()
```
