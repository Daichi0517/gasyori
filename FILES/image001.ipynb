```python
#ライブラリをインポート
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow

#画像の読み込み
img = cv2.imread('imori.jpg')
img_noise = cv2.imread('imori_noise.jpg')
#BGRとRGBを変換
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#画像を表示する
cv2_imshow(img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
